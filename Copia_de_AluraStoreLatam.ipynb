### Importación de datos



```python
# Importación de librerías y lectura de conjunto de datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación

```python
# Cálculo ingresos totales de cada tienda
precio_total_T1 = tienda['Precio'].sum()
precio_total_T1
```

```python
precio_total_T2 = tienda2['Precio'].sum()
precio_total_T2
```

```python
precio_total_T3 = tienda3['Precio'].sum()
precio_total_T3
```

```python
precio_total_T4 = tienda4['Precio'].sum()
precio_total_T4
```

```python
#Gráfico de los ingresos totales por tienda
ingresos_totales = {
    'Tienda 1': precio_total_T1,
    'Tienda 2': precio_total_T2,
    'Tienda 3': precio_total_T3,
    'Tienda 4': precio_total_T4
}

df_ingresos = pd.Series(ingresos_totales)

plt.figure(figsize=(10, 6))
df_ingresos.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales (COP)')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#Cálculo de ventas por categoría
ventas_por_categoria_T1= tienda.groupby('Categoría del Producto').size().reset_index(name='Ventas')
print('Ventas por categoría en Tienda 1:')
ventas_por_categoria_T1
```

```python
#Resumen de la categoría más y menos vendida de cada tienda
stores_category_sales = {
    'Tienda 1': ventas_por_categoria_T1,
    'Tienda 2': ventas_por_categoria_T2,
    'Tienda 3': ventas_por_categoria_T3,
    'Tienda 4': ventas_por_categoria_T4
}

for store_name, df_category_sales in stores_category_sales.items():
    if not df_category_sales.empty:
        most_sold_category = df_category_sales.loc[df_category_sales['Ventas'].idxmax()]
        least_sold_category = df_category_sales.loc[df_category_sales['Ventas'].idxmin()]

        print(f"\n{store_name}:")
        print(f"  Categoría más vendida: '{most_sold_category['Categoría del Producto']}' con {most_sold_category['Ventas']} ventas.")
        print(f"  Categoría menos vendida: '{least_sold_category['Categoría del Producto']}' con {least_sold_category['Ventas']} ventas.")
    else:
        print(f"\n{store_name}: No hay datos de ventas por categoría disponibles.")
```
